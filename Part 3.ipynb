```python
# imports 

import pandas as pd
import numpy as np
```

```python
# Basics DF

basics = pd.read_csv('Data/title_basics.csv.gz')
basics.head()
```

```python
# drop columns you do not want in new table

basics = basics.drop(columns = ['titleType', 'originalTitle', 'isAdult', 'endYear'])
basics.head()
```

```python
## create a col with a list of genres

basics['genres_split'] = basics['genres'].str.split(',')
basics
```

```python
exploded_genres = basics.explode('genres_split')
exploded_genres
```

```python
unique_genres = sorted(exploded_genres['genres_split'].unique())
unique_genres
```

```python
# save tconst and genre split as new df

title_genres = exploded_genres[['tconst', 'genres_split']].copy()
title_genres.head()
```

```python
## Making the genre mapper dictionary

genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
## Make new integer genre_id and drop string genres

title_genres['genre_id'] = title_genres['genres_split'].map(genre_map)
title_genres = title_genres.drop(columns='genres_split')

title_genres.head()
```

```python
# Make df mannually with named cols from the .keyd and .values

genre_lookup = pd.DataFrame({'Genre_Name': genre_map.keys(),
                            'Genre_ID': genre_map.values()})

genre_lookup.head()
```
